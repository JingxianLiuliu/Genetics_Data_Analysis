```python
import pandas as pd

# Read in files
fn_fbg = 'Patient_FBG.csv'
fn_sex = 'Patient_sex.csv'
df_fbg = pd.read_csv(fn_fbg)
df_sex = pd.read_csv(fn_sex)

# Add SEX information to FBG dataframe
df_merged = df_fbg.merge(df_sex, left_on='PATIENT_ID', right_on='PATIENT_ID')

# ------------- Diabetic -------------
# (3 conditions)
# year_1>125
# year_1 is NA, year_2>125
# year_1 and year_2 are NAs, year_3>125
mask_diabetic = (df_merged['FBG_YEAR_1'] > 125) | \
                ((df_merged['FBG_YEAR_1'].isna()) & (df_merged['FBG_YEAR_2'] > 125)) | \
                ((df_merged['FBG_YEAR_1'].isna()) & (df_merged['FBG_YEAR_2'].isna()) & (df_merged['FBG__YEAR_3'] > 125))

# ------------- Transitional -------------
# If year_1 healthy, then compare year_2 or year_3 to 125
# If year_1 is NA and year_2 is healthy, then compare year 3 to 125
mask_transitional = ((df_merged['FBG_YEAR_1']<126) & ((df_merged['FBG_YEAR_2'] > 125) | (df_merged['FBG__YEAR_3'] > 125))) | \
                    ((df_merged['FBG_YEAR_1'].isnull()) & (df_merged['FBG_YEAR_2'] < 126) & (df_merged['FBG__YEAR_3'] > 125))

# ------------- NA -------------
number_of_NA = df_merged.iloc[:,1:4].isna().all(axis=1).sum()
print('Number of NA is', number_of_NA)

# ------------- Healthy -------------
# Drop patients wih NA status first, since I do not have a mask for NA status
df_merged.dropna(thresh=3, inplace=True)
# If an individual is not diabetic or transitional, then he/she is healthy
mask_healthy =  (mask_diabetic | mask_transitional) == False

# ------------- Print out results (or write to a file) -------------
print('Number of diabetic male is', df_merged.loc[mask_diabetic & (df_merged['SEX']=='M')].shape[0])
print('Number of diabetic female is', df_merged.loc[mask_diabetic & (df_merged['SEX']=='F')].shape[0])
print('Number of transitional male is', df_merged.loc[mask_transitional & (df_merged['SEX']=='M')].shape[0])
print('Number of transitional female is', df_merged.loc[mask_transitional & (df_merged['SEX']=='F')].shape[0])
print('Number of healthy male is', df_merged.loc[mask_healthy & (df_merged['SEX']=='M')].shape[0])
print('Number of healthy female is', df_merged.loc[mask_healthy & (df_merged['SEX']=='F')].shape[0])
```
