# Minor allele frequency (MAF)

```python
# First, select specific data for calculation
# Then, Returning a list-like will result in a Series
maf= (df_ped.iloc[:, np.r_[2:102]]).apply(lambda x : x.value_counts(),axis='rows')

# Fill out the missing data--Fill NA/NaN values using the specified method
maf.fillna(value=0,inplace=True)  #inplace-if True, fill in-place.

#transpose the current dataframe
maf = pd.DataFrame(maf.values.T,columns=maf.index,index=maf.columns)
maf.head()

# Sum and calculate all of the allele
maf['0'] = 2*maf['00'] 
maf['*AG'] = maf['GAG'] 
maf['*AC'] = maf['AAC'] + 2*maf['ACAC'] 
maf['AAAAC'] = maf['AAAACA'] + 2*maf['AAAACAAAAC'] 

maf['*CT'] = maf['CCT'] + 2*maf['CTCT'] 
maf['*TC'] = maf['CTC'] + 2*maf['TCTC'] + maf['TTC']
maf['*TG'] = maf['TTG'] + 2*maf['TGTG'] 

maf['A'] = 2*maf['AA'] + maf['AAAACA'] + maf['AAC'] + maf['AG'] + maf['AT'] + maf['CA'] +maf['GA'] + maf['TA'] + maf['AC']
maf['G'] = maf['AG'] + maf['CG'] + maf['GA'] + maf['GAG'] + maf['GC'] + 2*maf['GG'] + maf['GT'] + maf['TG'] 
maf['T'] = maf['AT'] + maf['GT'] + maf['CT'] + maf['TA'] + maf['TC'] + maf['TG'] + 2*maf['TT'] + maf['TTC'] + maf['TTG']
maf['C'] = maf['AC'] + maf['CA'] + 2*maf['CC'] + maf['CT'] + maf['TC'] + maf['CTC'] + maf['CCT'] + maf['CG'] + maf['GC'] 
maf.head()
```

```python
#select the columns of the allele above, respectively
maf_allele = maf.iloc[:, np.r_[30:40]]
maf_allele.head()
```

```python
#To exclude the missing values'inluence, firstly, we replace the missings with 300(any # larger than 200)
maf_allele.replace(0,300,inplace = True)
#maf_allele.head()
maf_allele['min_allele'] = maf_allele.idxmin(axis=1)
maf_allele['maf'] = (maf_allele.min(axis=1))/200
maf_allele.head()
```

```python
# select/leave the columns that we need for print-- the least often occurring allele and the maf
maf_print = maf_allele.iloc[:, np.r_[10:12]]
maf_print.head()

# Satisfy the 0.01 thresholds for Minor allele frequency
maf_missings_excluded = maf_print[(maf_print['maf']<=0.01)]
maf_missings_excluded.head()
```

```python
# write the answer to the assigned document
f = open('variant_maf_excluded_Jingxian.txt', 'w')
print(maf_missings_excluded, file = f)
f.close()
```
